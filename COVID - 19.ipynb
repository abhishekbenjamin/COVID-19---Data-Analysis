# Null Value Check & Missing Value Imputation

```python
df.isnull().sum()
```

```python
df['Total Cases'] = df['Total Confirmed cases (Indian National)'] + df['Total Confirmed cases ( Foreign National )']

df.head(3)
```

### Overall Cases Reported in India so Far

```python
Overall_Cases = df['Total Cases'].sum()

print('Overall Cases Reported in India: ', Overall_Cases)
```

```python
df['Active Cases'] = df['Total Cases'] - (df['Cured'] + df['Death'])

df.head()
```

```python
df.style.background_gradient(cmap='Reds')
```

# Grouping Data State-Wise

```python
Total_Active_Cases = df.groupby(['Name of State / UT'])['Total Cases'].sum().sort_values(ascending = False).to_frame()

Total_Active_Cases.style.background_gradient(cmap = 'Blues')
```

   # Visualization

```python
plt.figure(figsize=(20,18))
df.plot(kind = 'barh', x= 'Name of State / UT', y = 'Total Cases')
```

```python
fig=plt.figure(figsize=(20,10),dpi=200)

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.bar(df['Name of State / UT'],df['Total Cases'])
axes.set_title("Total Cases in India")
axes.set_xlabel("Name of State / UT")
axes.set_ylabel("Total Cases")

plt.show()
```
